```python
# 此文件运行将语法分析器生成的x-out.txt文件，转化为可以被graphviz绘图的.dot文件
# 注意次文件运行会将x-out.txt反序，因此要先运行run.sh重新生成，再运行此文件

# 指定文件路径
input_filename = './test/2-out.txt'
output_filename = './test/2-out.dot'

# 反序文件行
def reverse_lines(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()

    reversed_lines = lines[::-1]

    with open(filename, 'w') as file:
        file.writelines(reversed_lines)

#解析成.dot文件
def parse_to_dot(input_file, output_file):
    with open(input_file, 'r') as file:
        lines = file.readlines()

    with open(output_file, 'w') as file:
        file.write('digraph " "{\n')
        file.write('node [shape = record,height=.1]\n')

        cnt = 0
        node_map = {}
        map = {}

        for line in lines:
            parts = line.strip().split(' -> ')
            source_node = parts[0]
            target_node = parts[1]

            # node_map 存储每个节点对应的部分的编号，比如node_map[i] = [200,300,100]

            # source分析，创建字典
            source = source_node.strip().split(':')
            map[source[0]] = source[1]
            # target分析，创建字典
            target_list = target_node.strip().split(' ')
            target_vec = []
            for t in target_list:
                target = t.strip().split(':')
                map[target[0]] = target[1]

            
            if cnt == 0: #第一次进来，需要创建两个节点
                # source节点创建
                file.write(f'node{cnt}[label = "<f0> {source[1]}"];\n')

                node_map_list_source = []
                node_map_list_source.append(source[0])
                node_map[cnt] = node_map_list_source
                cnt += 1

                # target节点创建
                file.write(f'node{cnt}[label = "')
                i = 0
                node_map_list_target = []
                for t in target_list:
                    target = t.strip().split(':')
                    if i>0:
                        file.write(f'|<f{i}> {target[1]}')
                    else:
                        file.write(f'<f{i}> {target[1]}')
                    node_map_list_target.append(target[0])
                    i += 1

                node_map[cnt] = node_map_list_target
                cnt += 1
                file.write(f'"];\n')

                #链接两个节点
                file.write(f'"node0":f0->"node1";\n')


            # 第二次之后进来，只需找到source节点创建的node，然后target节点创建
            # 在node_map中找到source节点创建的node，得到该node的编号source_node和source_f
            else:
                list1 = []
                source_node = 0
                for key in node_map.keys():
                    if  source[0] in node_map[key]:
                        source_node = key
                        break
                
                list1 = node_map[source_node]
                source_f = list1.index(source[0])
                
                # target节点创建
                file.write(f'node{cnt}[label = "')
                i = 0
                node_map_list_target = []
                for t in target_list:
                    target = t.strip().split(':')
                    if i>0:
                        file.write(f'|<f{i}> {target[1]}')
                    else:
                        file.write(f'<f{i}> {target[1]}')
                    node_map_list_target.append(target[0])
                    i += 1

                node_map[cnt] = node_map_list_target
                cnt += 1
                file.write(f'"];\n')

                #链接两个节点
                file.write(f'"node{source_node}":f{source_f}->"node{cnt-1}";\n')
        file.write('}')

    #print(map)
    print(node_map)

# 反序文件行
reverse_lines(input_filename)

# 解析-out.txt文件为.dot文件
parse_to_dot(input_filename, output_filename)
```
